## Modify case

Basic knowledge of pwoer flow simulaiton:

1) PQ bus: give active power p, reactive power q, sovle voltage amplitude and angle

2) PV bus: give active power p, votlage amplitude, solve reactive pwoer q and votlage angle

3) Slack: give voltage amplitue = 1 and voltage angle = 0, solve active power p and reactive power q

### Modify power flow case

```python
# Modify load data

# ------- Scale all loads ---------
ssa.PQ.set(
            src='p0', 
            idx= ssa.PQ.idx.v, 
            attr='v',
            value= 1.1 * ssa.PQ.p0.v
        )

ssa.PQ.set(
            src='q0', 
            idx= ssa.PQ.idx.v, 
            attr='v',
            value= 1.1 * ssa.PQ.q0.v
        )
```

```python
# Change specific load based on idx
# Note: check the case (excel) file to get the correct idx
ssa.PQ.set(
            src='p0', 
            idx= ['PQ_1', 'PQ_10'], 
            attr='v',
            value= np.array([8, 4])
        )

ssa.PQ.set(
            src='q0', 
            idx= ['PQ_1', 'PQ_10'], 
            attr='v',
            value= np.array([2, 1])
        )
```

```python
# change power reference of PV/Slack bus 

# PV bus
# Note: check the case (excel) file to get the correct idx
# If the idx is number, then use idx= [1, 2] instead of idx= ['1', '2']
ssa.PV.set(
            src='p0', 
            idx= [1, 2], 
            attr='v',
            value= np.array([5, 7])
        )

ssa.PV.set(
            src='q0', 
            idx= [1, 2], 
            attr='v',
            value= np.array([2, 2.5])
        )
```

```python
# Slack bus
# Note: check the case (excel) file to get the correct src and idx
ssa.Slack.set(
            src='v0', 
            idx= 10, 
            attr='v',
            value= np.array([1])
        )

ssa.Slack.set(
            src='a0', 
            idx= 10, 
            attr='v',
            value= np.array([0])
        )
```

### Add new components

Note: Adding new components can only be done before setting up the case '''ssa.setup()'''

So, we will need to load a new case and work on adding new components

```python
ssa = andes.load(
                andes.get_case('ieee39/ieee39.xlsx'),
                setup=False,     # If setup=False, need to call ss.setup() before running the simulation
                no_output=True, 
                default_config=False
            )
```

```python
# Add new load
# Remove load can be done by setting the p0 and q0 to 0
ssa.add(model="PQ", param_dict=dict(bus=2, 
                                    idx="PQ_20", 
                                    p0=0.5, 
                                    q0=0.2
                                    ))
```

```python
# We can check the all the compoennts by printing the component as dataframe
# ssa.PQ.as_df()
```

```python
# Add new generator
# Remove generator can be done by setting the p0 and q0 to 0
ssa.add(model="PV", param_dict=dict(bus=15, 
                                    idx='15', 
                                    p0=0.5, 
                                    q0=0.2,
                                    pmax=5,
                                    pmin=0,
                                    qmax=2,
                                    qmin=-2,
                                    v0=1.0,
                                    ))
```

```python
# We can check the all the compoennts by printing the component as dataframe
# ssa.PV.as_df()
```

```python
ssa.setup()
ssa.PFlow.run()
```
