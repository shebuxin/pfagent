```python
# import the andes package
import andes
import pandas as pd
import numpy as np
```

```python
# load the IEEE 14-bus system
ssa = andes.load(
                andes.get_case('ieee14/ieee14_full.xlsx'),
                setup=True,     # If setup=False, need to call ss.setup() before running the simulation
                no_output=True, 
                default_config=False
            )
```

```python
# call the power flow calculation
ssa.PFlow.run()
```

```python
# Extract the results
ssa_res_gen = pd.DataFrame(columns=['name', 'p_mw', 'q_mvar', 'va_degree', 'vm_pu'])
ssa_res_gen['name'] = ssa.PV.as_df()['name']
ssa_res_gen['p_mw'] = ssa.PV.p.v * ssa.config.mva
ssa_res_gen['q_mvar'] = ssa.PV.q.v * ssa.config.mva
ssa_res_gen['va_degree'] = ssa.PV.a.v * 180 / np.pi
ssa_res_gen['vm_pu'] = ssa.PV.v.v

# Extract the slack bus results
ssa_res_slack = pd.DataFrame([[ssa.Slack.name.v[0], ssa.Slack.p.v[0] * ssa.config.mva,
                              ssa.Slack.q.v[0] * ssa.config.mva, ssa.Slack.a.v[0] * 180 / np.pi,
                              ssa.Slack.v.v[0]]],
                             columns=ssa_res_gen.columns,
                             )

# Combine the generator and slack bus results
ssa_res_gen = pd.concat([ssa_res_gen, ssa_res_slack]).reset_index(drop=True)

```
